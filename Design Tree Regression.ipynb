```python
from sklearn.model_selection import GridSearchCV

# Hyperparameter Tuning for Decision Tree Regressor using GridSearchCV
param_grid = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2']
}
```

```python
grid_search = GridSearchCV(DecisionTreeRegressor(random_state=42), param_grid, cv=5, n_jobs=-1, verbose=1, scoring='neg_mean_squared_error')
grid_search.fit(x_train, y_train)

# Best hyperparameters
print("Best Hyperparameters:", grid_search.best_params_)
```

```python
best_tree = grid_search.best_estimator_

# Predictions
y_pred_best = best_tree.predict(x_test)
```

```python
# Evaluation Metrics for Tuned Model
mse_best = mean_squared_error(y_test, y_pred_best)
mae_best = mean_absolute_error(y_test, y_pred_best)
r2_best = r2_score(y_test, y_pred_best)

print(f"Mean Squared Error (MSE) after tuning: {mse_best:.2f}")
print(f"Mean Absolute Error (MAE) after tuning: {mae_best:.2f}")
print(f"R-squared (R2) after tuning: {r2_best:.2f}")
```

```python
# Plotting the Tuned Decision Tree
plt.figure(figsize=(20, 10))
plot_tree(best_tree, filled=True, feature_names=feature_names, rounded=True)
plt.title("Tuned Decision Tree Diagram")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
importance_best = best_tree.feature_importances_
sns.barplot(x=importance_best, y=feature_names)
plt.title("Feature Importance After Tuning")
plt.show()
```
