```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import seaborn as sns

# Load your dataset
# Make sure to replace 'your_dataset.csv' with the actual file path or DataFrame
df = pd.read_csv("./data/normalized_data")

# Assume 'claim_amount' is the target variable and all other columns are features
X = df.drop(columns=['CLM_AMT'])  # Features
y = df['CLM_AMT']  # Target variable

# Optional: Log transformation of target variable if needed
y_log = np.log1p(y)

# Split the data into training and testing sets
X_train, X_test, y_train_log, y_test_log = train_test_split(X, y_log, test_size=0.2, random_state=42)

# Scale the features (if necessary)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train the Decision Tree Regressor
model = DecisionTreeRegressor(random_state=42)
model.fit(X_train_scaled, y_train_log)

# Make predictions on the test set
y_pred_log = model.predict(X_test_scaled)

# Convert predicted log values back to original scale
y_pred_original = np.exp(y_pred_log)

# Convert actual log values back to original scale
y_test_original = np.exp(y_test_log)

# Evaluate the model
mse = mean_squared_error(y_test_original, y_pred_original)
mae = mean_absolute_error(y_test_original, y_pred_original)

print(f'Mean Squared Error: {mse:.2f}')
print(f'Mean Absolute Error: {mae:.2f}')

# Plot Actual vs Predicted Claim Amounts (Original Scale)
plt.figure(figsize=(12, 6))
plt.scatter(y_test_original, y_pred_original, alpha=0.5, label='Predicted vs Actual')
plt.plot([0, max(y_test_original)], [0, max(y_test_original)], color='red', linestyle='--', label='Perfect Prediction')  # Diagonal line
plt.title('Actual vs. Predicted Claim Amounts (Original Scale)')
plt.xlabel('Actual Claim Amount')
plt.ylabel('Predicted Claim Amount')
plt.xlim([0, max(y_test_original) + 10000])  # Adjust limits for better visibility
plt.ylim([0, max(y_pred_original) + 10000])  # Adjust limits for better visibility
plt.legend()
plt.grid()
plt.show()

# Optional: Visualize distribution of claim amounts
plt.figure(figsize=(12, 6))
sns.histplot(df['claim_amount'], bins=50, kde=True)
plt.title('Distribution of Claim Amounts')
plt.xlabel('Claim Amount')
plt.ylabel('Frequency')
plt.show()
```
